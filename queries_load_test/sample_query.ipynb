```python
import pandas as pd

parquet_path='../data_preparation/parquet/'
```

```python
title_full=pd.concat([pd.read_parquet(parquet_path+'title.basics.parquet'),pd.read_parquet(parquet_path+'title.episodes.parquet')]).drop(columns=['titleId'])
```

```python
title_full
```

```python
title_akas=pd.read_parquet(parquet_path+'title.akas.parquet')
```

```python
title_akas
```

```python
to_query=title_akas.join(title_full,
                        on='titleId',
                        how='inner',
                        lsuffix='_akas',
                        rsuffix='_full')[["titleId","nameLower","votes"]].sort_values(["votes","titleId"],ascending=False)
```

```python
to_query.sample(n=10000, random_state=1, replace=True, weights=to_query["votes"])[["nameLower","titleId"]].to_csv('query.csv', index=False, header=False)
```
